```python
from sklearn.datasets import load_digits # 패키지 임포트
import matplotlib.pylab as plt
mpl.rc('font', family='NanumBarunGothic')

digits = load_digits() # 데이터 로드
samples = [0, 10, 20, 30, 1, 11, 21, 31] # 선택된 이미지 번호
d = []
for i in range(8):
    d.append(digits.images[samples[i]])
    
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False);
    plt.xticks([])
    plt.yticks([])
    plt.title('image {}'.format(i+1))
plt.suptitle('숫자 0과 1 이미지')
plt.tight_layout()
plt.show()
```

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1))
    
plt.figure(figsize=(8,3))
for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(v[i], aspect=0.4, interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title('벡터 {}'.format(i+1))
plt.suptitle('벡터화된 이미지', y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

```python
from scipy import misc # 패키지 임포트
img_rgb = misc.face() # 컬러이미지 로드
img_rgb.shape # 데이터의 모양
```

```python
plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray) # 컬러이미지
plt.axis(False)
plt.title('RGB컬러이미지')

plt.subplot(222)
plt.imshow(img_rgb[:,:,0], cmap=plt.cm.gray) # Red 채널
plt.axis(False)
plt.title('Red 채널')

plt.subplot(223)
plt.imshow(img_rgb[:,:,1], cmap=plt.cm.gray) # Green 채널
plt.axis(False)
plt.title('Green 채널')

plt.subplot(224)
plt.imshow(img_rgb[:,:,2], cmap=plt.cm.gray) # Blue 채널
plt.axis(False)
plt.title('Blue 채널')
```
