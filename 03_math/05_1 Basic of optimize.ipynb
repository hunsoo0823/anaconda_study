```python
import scipy as sp
import scipy.optimize
```

```python
# 목적함수 재정의
def f1(x):
    return (x-2) ** 2 + 2

x0 = 0 # 초기값
result = sp.optimize.minimize(f1, x0) # optimization
print(result)
```

```python
def f1p(x):
    return 2 * (x-2)

result = sp.optimize.minimize(f1, x0, jac=f1p)
print(result)
```

```python
# 2차원 목적함수 재정의(벡터 입력을 받도록)
def f2(x):
    return (1 - x[0])**2 + 100.0 * (x[1] - x[0]**2)**2

x0 = (-2,-2)
result = sp.optimize.minimize(f2, x0)
print(result)
```

```python
x0 = (1,1)
result = sp.optimize.minimize(f2, x0)
print(result)
```

```python
def f1p(x):
    return 2 * (x-2)

result = sp.optimize.minimize(f2, x0, jac=f1p)
print(result)
```
