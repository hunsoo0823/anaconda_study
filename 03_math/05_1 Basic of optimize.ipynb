```python
import scipy as sp
import scipy.optimize
```

```python
# 목적함수 재정의
def f1(x):
    return (x-2) ** 2 + 2

x0 = 0 # 초기값
result = sp.optimize.minimize(f1, x0) # optimization
print(result)
```

```python
def f1p(x):
    return 2 * (x-2)

result = sp.optimize.minimize(f1, x0, jac=f1p)
print(result)
```

```python
# 2차원 목적함수 재정의(벡터 입력을 받도록)
def f2(x):
    return (1 - x[0])**2 + 100.0 * (x[1] - x[0]**2)**2

x0 = (-2,-2)
result = sp.optimize.minimize(f2, x0)
print(result)
```

```python
x0 = (1,1)
result = sp.optimize.minimize(f2, x0)
print(result)
```

```python
def f1p(x):
    return 2 * (x-2)

result = sp.optimize.minimize(f2, x0, jac=f1p)
print(result)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f1(x):
    return (x-2) ** 2 + 2

xx = np.linspace(-1,4,100)
plt.plot(xx, f1(xx))
plt.plot(2,2,'ro', markersize=10)
plt.ylim(0,10)
plt.xlabel("x")
plt.ylabel("$f_1(x)$")
plt.title("1Dimension objective function")
plt.show()
```

```python
def f2(x,y):
    return (1-x)**2 + 100.0 * (y - x**2)**2

xx = np.linspace(-4, 4, 800)
yy = np.linspace(-3, 3, 600)
X, Y = np.meshgrid(xx, yy)
Z = f2(X,Y)

levels = np.logspace(-1, 3, 10)
plt.contourf(X,Y,Z, alpha=0.2, levels=levels)
plt.contour(X,Y,Z, colors="gray", levels=[0.4,3,15,50,150,500,1500,5000])
plt.plot(1,1,'ro',markersize=10)

plt.xlim(-4,4)
plt.ylim(-3,3)
plt.xticks(np.linspace(-4,4,9))
plt.yticks(np.linspace(-3,3,7))
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("2Dimension Rosenblock $f(x,y)$")
plt.show()
```
