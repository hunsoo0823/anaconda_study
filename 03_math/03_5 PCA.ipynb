```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
N = 10 # 앞의 10송이만 선택 
X = iris.data[:N, :2] # 꽃받침의 길이와 꽃받침 폭만 선택

plt.plot(X.T, 'o:')
plt.xticks(range(4))
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title("Iris")
plt.legend(["specimen {}".format(i+1) for i in range(N)])
plt.show()
```
