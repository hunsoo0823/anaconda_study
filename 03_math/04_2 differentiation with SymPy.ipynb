```python
import sympy
```

```python
sympy.init_printing(use_latex='mathjax')
```

```python
x = sympy.symbols('x')
x
```

```python
type(x)
```

```python
f = x * sympy.exp(x)
f
```

```python
sympy.diff(f)
```

```python
sympy.simplify(sympy.diff(f))
```

```python
x, y = sympy.symbols('x y')
f = x ** 2 + 4 * x * y + 4 * y * y ** 2
f
```

```python
sympy.diff(f, x)
```

```python
sympy.diff(f, y)
```

```python
x, mu, sigma = sympy.symbols('x mu sigma')
f = sympy.exp((x - mu) ** 2 / sigma ** 2)
f
```

```python
sympy.diff(f, x)
```

```python
sympy.simplify(sympy.diff(f, x))
```

```python
sympy.diff(f, x, x)
```
