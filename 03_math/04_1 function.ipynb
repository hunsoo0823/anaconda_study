```python
def heaviside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        return 1.0 if x>=0 else 0.0
    
heaviside_step(-0.0001), heaviside_step(0), heaviside_step(0.0001),  
```

```python
def f(x):
    return x**3 - 3 * x**2 + x
```

```python
x = np.linspace(-1, 3, 9)
x
```

```python
y = f(x)
y
```

```python
import matplotlib.pylab as plt
```

```python
plt.plot(x, y, 'ro-')
plt.xlim(-2,4)
plt.title("function  $f(x) = x^3 = 3x^2 +x$ graph")
plt.xlabel("x")
plt.xlabel("y")
plt.xticks(np.arange(-1,5))
plt.yticks(np.arange(-5,4))
plt.show()
```

```python
x = np.linspace(-1, 3, 400)
y = f(x)
plt.plot(x,y)

plt.xlim(-2, 4)
plt.title("function  $f(x) = x^3 = 3x^2 +x$ graph")
plt.xlabel("x")
plt.xlabel("y")
plt.xticks(np.arange(-1,5))
plt.yticks(np.arange(-5,4))

plt.show()
```

```python
x = np.linspace(-2,2,400)
y = np.sign(x)
plt.plot(x,y)

plt.xlim(-3, 4)
plt.title("sign function")
plt.xlabel("x")
plt.xlabel("y")
plt.xticks(np.arange(-1,5))
plt.yticks(np.arange(-5,4))

plt.show()
```
