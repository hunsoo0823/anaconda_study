```python
A = np.array([[60.,30.,20.],
              [30.,20.,15,],
              [20.,15.,12.]])

w , V  = np.linalg.eig(A)
w1, w2, w3 = w
v1 = V[:, 0:1]
v2 = V[:, 1:2]
v3 = V[:, 2:3]
A1 = v1 @ v1.T
A2 = v2 @ v2.T
A3 = v3 @ v3.T
```
