```python
import scipy as sp
import scipy.optimize
import numpy as np
```

```python
def f1array(x):
    #return x[0] ** 2 + x[1] ** 2
    return x[0] + x[1]
```

```python
def eq_constarint(x):
    #return x[0] + x[1] - 1
    return x[0] ** 2 + x[1] ** 2 - 1
```

```python
sp.optimize.fmin_slsqp(f1array, np.array([1,1]), eqcons = [eq_constarint])
```

```python
def f2(x):
    return np.sqrt((x[0] - 4) ** 2 + (x[1] -2 ) ** 2)

# 제한 조건 상수
k = 10

def ieq_constraint(x):
    return np.atleast_1d(k - np.sum(np.abs(x)))
```

```python
sp.optimize.fmin_slsqp(f2, np.array([0,0]), ieqcons = [ieq_constraint])
```
