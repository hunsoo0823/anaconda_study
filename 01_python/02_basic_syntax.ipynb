### 6. 연산자
    - 산술연산자 : +, -, *, /, //, %, **
    - 할당연산자 : 변수에 누적시켜서 연산 : +=, //= , **=
    - 비교연산자 : >, <, ==, !=, <= >= : 결과로 True, False
    - 논리연산자 : True,False 를 연산 : or, and, not
    - 멤버연산자 : 특정 데이터가 있는지 확인할때 사용 : not, in, in

```python
# 산술 연산
(1+4)/2**2
```

```python
# 할당 연산
a = 10
# a = a + 10
# a = a + 10
a += 10
a += 10
a
```

```python
# 비교연산
print(a,b)
a < b, a==b, a!= b
```

```python
# 솔루션을 리스트로 작성
# 질문 입력 받음
# 솔루션을 갯수에 맞게 랜덤한 index 정수 값을 생성
# index 해당하는 솔루션 리스트와 데이터를 출력
```

```python
solutions = [
    "무엇을 하던 잘 안될것 이다.",
    "생각지도 않게 좋은 일이 생길것이다.",
    "무엇을 상상하던 그 이상이다."
]

# 질문을 입력 받음
input("질문을 입력하세요. : ")
# 솔루션의 갯수에 맞게 랜덤한 inedx 정수 값을 생성
idx = random.randint(0, len(solutions)-1)

# index에 해당하는 솔루션 리스트의 데이터를 출력
solutions[idx]
```
