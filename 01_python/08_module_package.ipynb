### summart
- 상속 : 다른 클래스를 받아서 기능을 추가해서 새로운 클래스를 만드는 방법
- super : 부모클래스에서 특정 함수의 코드를 가져오는 방법
- getter, setter : 클래스로 만들어진 객체에 변수값을 수정하거나 출력할때 특정 함수를 통해서 수정하고 출력하는 방법
- non public(private) : mangling (__), `_(클래스명)`이 붙은 변수로 객체를 생성할때 변경이 되어서 생성
- is a / has a : 클래스를 설계하는 방법 
- magic(spacial) method
    - 비교 
        - `__eq__`(==), `__ne__`(!=), `__lt__(<)`, `__gt__(>)`, `__le__`(<=), `__ge__(>=)`
    - 연산
        - `__da__`(+), `__sub__`(-), `__mul__`(`*`), `__truediv__`(/)
        - `__floordiv__`(//), `__mod__`(%), `__pow__`(**)
    - 그외 : `__repr__`, `__str__`

```python
# Integer 객체
class Interger:
    def __init__(self, number):
        self.number = number
    
    def __add__(self, obj):
        return self.number + obj.number

    def __str__(self):
        return str(self.number)
    
    def __repr__(self):
        return str(self.number)
```

```python
num1 = Interger(1)
num2 = Interger(2)
```

```python
num1 + num2
```

```python
# 클래스 계좌 클래스 만들기 
# 변수 : 자산(asset), 이자율(interest)
# 함수 : 인출(draw), 입금(insert), 이자추가(add_interst)
# 인출시 자산이상의 돈을 인출할 수 없습니다.
```

```python
class account:
    def __init__(self, asset, interest):
        self.asset = asset
        self.interest = interest
    def draw(self, draw_money):
        if draw_money > self.asset:
            print("자산이상의 돈을 인출할 수 없습니다.")
            
        else:
            self.asset -= draw_money
            print("{}원의 자산을 인출하셨습니다.".format(draw_money))
            print("{}원의 잔액이 남았습니다.".format(self.asset))
            
    def insert(self, insert):
        self.asset += insert
        print("{}원의 자산을 입금하셨습니다.".format(insert))
        print("{}원의 잔액이 남았습니다.".format(self.asset))
        
    def add_interest(self, interest):
        self.interest += interest
        print("현재 이자율은 {}% 입니다.".format(self.interest))
```

```python
acc1 = account(5000, 20)
```

```python
acc1.draw(100)
```

```python
acc1.insert(200)
```

```python
acc1.add_interest(2)
```
