### 예외처리
- 코드를 실행중에 에러가 발생한 경우 에러를 처리하는 방법
- try, except, finally, rais

```python
# try, except : 에러가 발생해도 코드의 실행을 계속하고 싶을때
ls = [1,2,3] 
```

```python
print(ls[3])
print("Done!")
```

```python
try:
    print(ls[3])
except Exception as e:
    print("error")
    print(e)
print("Done")
```

```python
# finally : try,except 구문이 실행된 후 finally 구문이 실행
try:
    1/0
except:
    print("error")
finally:
    print("Done!")
```

```python
# raise : 강제로 에러를 발생시키는 명령
```

```python
# finally : try,except 구문이 실행된 후 finally 구문이 실행
try:
    1/0
except Exception as e:
    print("error")
    raise(e)

print("Done!")
```
