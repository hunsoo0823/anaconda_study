```python
money = 12000

if money >= 10000:
    print("택시를 타고 집에 갑니다.")
    
if money < 10000:
    print("걸어서 집에 갑니다.")
```

```python
money = 5000

if money >= 10000:
    print("택시를 타고 집에 갑니다.")
elif money >= 5000:
    print("광역버스를 타고 집에 갑니다.")
elif money >= 2000:
    print("일반버스를 타고 집에 갑니다.")
else:
    print("걸어서 집에 갑니다.")
```

```python
# 계좌에 10000원이 들어 있습니다.
# 인출 금액을 입력 받습니다.
# 인출 금액이 계좌에 있는 금액보다 크면 "인출이 불가능합니다." 출력
# 인출 금액이 계좌에 있는 금액보다 작으면 "인출 되어있습니다" 출력
# 마지막 현재 계좌의 잔액을 출력
```

```python
account = 10000
```

```python
data = int(input("draw money : "))

if data > account:
    print("인출이 불가능합니다."+str(data-account)+"원이 부족합니다.")
else:
    print(str(data)+"원이 인출 되어있습니다.")
    account -= data

print("현재 잔액은 "+str(account)+"원 입니다.")
```

```python
# string 데이터 타입의 format 함수
print("현재 잔액은 "+str(account)+" 원 입니다.")
print("현재 잔액은 {} 원 입니다. 인출금액은 {}입니다.".format(account, data))
print("현재 잔액은 {data1} 원 입니다. 인출금액은 {data2}입니다.".format(data2=data,data1=account))
```

#### 삼항 연산자
- 간단한 if, else 구문을 한줄의 코드로 작성
- (true) if (condition) else (False)

```python
# data 변수에 0이면 "zero" 출력, 아니면 "not zero" 출력
data = 1
if data:
    print("not zero")
else:
    print("zero")
```

```python
data = 1
result = "not zero" if  data else "zero"
result
```
