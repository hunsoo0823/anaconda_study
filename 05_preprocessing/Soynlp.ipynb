```python
from nltk.stem import PorterStemmer , LancasterStemmer

st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ["fly", "flies", "flying", "flew", "flown"]

print("Porter Stremmer : ",[st1.stem(w) for w in words])
print("Lancatser Stremmer : ",[st2.stem(w) for w in words])
```

```python
from nltk.stem import WordNetLemmatizer

Im = WordNetLemmatizer()

[Im.lemmatize(w, pos="v") for w in words]
```

```python
from nltk.tag import pos_tag
sentense = "Emma refused to permit us to obtain the refuse permit"
tagged_list = pos_tag(word_tokenize(sentense))
tagged_list
```

```python
def tokenizser(doc):
    return ["/".join(p) for p in tagged_list]

tokenizser(sentense)
```

```python
from nltk import FreqDist

stopwords = ["Mr.","Mrs.","Miss","Mr","Mrs","Dear"]
emma_tokens = pos_tag(word_tokenize(emma_raw))
names_list = [t[0] for t in emma_tokens if t[1]== "NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N() , fd_names["Emma"], fd_names.freq("Emma")
```

```python
fd_names.most_common(5)
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```

```python
sense_raw = nltk.corpus.gutenberg.raw("austen-sense.txt")
```

```python
from nltk import FreqDist

stopwords = ["Mr.","Mrs.","Miss","Mr","Mrs","Dear"]
sense_tokens = pos_tag(word_tokenize(sense_raw))
names_list = [t[0] for t in sense_tokens if t[1]== "NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```
