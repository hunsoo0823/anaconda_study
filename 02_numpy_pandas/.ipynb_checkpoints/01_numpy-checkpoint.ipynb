### Numpy
- 데이터는 행열로 표현
- 행열 데이터 빠르게 계산을 해야합니다.
- 행렬 데이터 생성, 수정, 계산등을 빠르게 처리해주는 패키지
- 특징
    - C,C++,포트란으로 작성
    - 선형대수학을 빠르게 연산
        - 스칼라, 벡터, 메트릭스

```python
import numpy as np
```

```python
# 행렬 데이터 생성
# ndarray : 한가지 데이터 타임만 값으로 사용이 가능
```

```python
array = np.array([1,2,3])
type(array), array
```

```python
array2 = np.array(
    [[1,2,3],
    [4,5,6]],
)

array2, array2.ndim, array2.shape
```

```python
# 행렬의 모양(shape) 변경하기
array2.reshape(3,2,1)
```

```python
# 행렬 데이터의 선택 : offset index : masking
```

```python
array2[1][:2]
```

```python
array2[1,2] #array2[1][2]
```
