### Numpy
- 데이터는 행열로 표현
- 행열 데이터 빠르게 계산을 해야합니다.
- 행렬 데이터 생성, 수정, 계산등을 빠르게 처리해주는 패키지

```python
import numpy as np
```

```python
array = np.array([1,2,3])
type(array), array
```
